```python
import pyodbc
import pandas as pd
```

```python
constants = { 
    'query' : """
                SELECT [dbo].[Users].Openness, [dbo].[Users].Conscientiousness, [dbo].[Users].Extraversion, [dbo].[Users].Agreeableness, [dbo].[Users].Neuroticism,
                [dbo].[Books].CommunalCoef, [dbo].[Books].AesthetiCoef, [dbo].[Books].DarkCoef, [dbo].[Books].ThrillingCoef, [dbo].[Books].CerebralCoef 
                FROM [dbo].[Users] JOIN [dbo].[Favourite] ON [dbo].[Favourite].userID = [dbo].[Users].ID
                LEFT JOIN [dbo].[Books] ON [dbo].[Favourite].bookID = [dbo].[Books].ID;
              """,
    'connection_string' : "Driver={SQL Server Native Client 11.0};"
                          "Server=bookshelf-sv.database.windows.net;"
                          "Database=bookshelf_db;"
                          "UID=adminlogin;"
                          "PWD=kartoffelpuree27AZURE;"
}
```

```python
# Retrieves data from the database and returns it 
# as a list of inputs and a list of outputs
def getData():
    # Connecting to the database and querying
    cnxn = pyodbc.connect(constants['connection_string'])
    cursor = cnxn.cursor()
    data = pd.read_sql(constants["query"], cnxn).values.tolist()
    
    # Separating data into input and output data
    inputs = []
    outputs = []
    for sample in data:
        inputs.append(sample[:5])
        outputs.append(sample[5:])
    
    return inputs, outputs
        
```

```python
inputs, outputs = getData()

print(*inputs, sep='\n')
print(*outputs, sep='\n')
```
